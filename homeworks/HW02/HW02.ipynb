```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../datasets/S02-hw-dataset.csv')
# Первичный осмотр
display(df.head(6))
df.info()
display(df.describe(include='all'))

# Доля пропусков по столбцам
print("Доля пропусков:")
print(df.isna().mean().round(3))

# Дубликаты строк
duplicates = df.duplicated().sum()
print(f"\nКоличество дублирующих строк: {duplicates}")

# Подозрительные значения
print("\nАномалии:")
invalid_age = df[(df['age'] < 0) | (df['age'] > 100)]
print(f"→ Возраст вне [0, 100]: {len(invalid_age)} записей (например: {invalid_age['age'].tolist()})")

neg_purchases = df[df['purchases'] < 0]
print(f"→ Отрицательные purchases: {len(neg_purchases)} записей ({neg_purchases['purchases'].tolist()})")

zero_purch_nonzero_rev = df[(df['purchases'] == 0) & (df['revenue'] > 0)]
print(f"→ purchases=0, но revenue>0: {len(zero_purch_nonzero_rev)} записей")

zero_rev_nonzero_purch = df[(df['revenue'] == 0) & (df['purchases'] > 0)]
print(f"→ revenue=0, но purchases>0: {len(zero_rev_nonzero_purch)} записей")

print("Распределение по странам:")
display(df['country'].value_counts())

grouped = df.groupby('country').agg(
    count_users=('user_id', 'count'),
    avg_age=('age', 'mean'),
    avg_purchases=('purchases', 'mean'),
    avg_revenue=('revenue', 'mean'),
    total_revenue=('revenue', 'sum')
).round(2)

print("\nАгрегаты по странам:")
display(grouped)


df_clean = df[(df['age'] <= 100) & (df['age'] >= 0) & (df['purchases'] >= 0)]

plt.figure(figsize=(6, 4))
plt.hist(df_clean['age'].dropna(), bins=8, color='lightcoral', edgecolor='black')
plt.title('Возраст пользователей')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('figures/img1.png')  
plt.show()

# 2. Боксплот revenue по странам
plt.figure(figsize=(6, 4))
df.boxplot(column='revenue', by='country', grid=False, patch_artist=True)
plt.title('Доход по странам')
plt.suptitle('')  
plt.xlabel('Страна')
plt.ylabel('Доход')
plt.tight_layout()
plt.savefig('figures/img2.png')
plt.show()

# 3. Scatter: purchases vs revenue
plt.figure(figsize=(6, 4))
plt.scatter(df['purchases'], df['revenue'], alpha=0.7, c='teal')
plt.title('Кол-во покупок/Доход')
plt.xlabel('Число покупок (purchases)')
plt.ylabel('Доход (revenue)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('figures/img3.png')
plt.show()
```

В столбце с возрастом найдено 2 пропущенных значения, а также имеется один полный дубликат строк. Присутствуют выбросы и некорректные записи: возраст 120 лет, отрицательное количество покупок (-1), а также две записи, где количество покупок равно нулю, но при этом указан положительный доход. По географическому распределению большинство пользователей представлено из России (14 человек), за которой следует Франция (11 человек). Анализ по странам показывает, что наибольший средний доход наблюдается у пользователей из Германии — около 550 единиц, несмотря на их малочисленность. В то же время самый низкий средний доход зафиксирован у Франции — приблизительно 700, хотя у французских пользователей отмечено наибольшее среднее количество покупок. Визуальный анализ выявляет положительную взаимосвязь между количеством покупок и доходом: с ростом покупок увеличивается и выручка. Однако на диаграмме рассеяния выделяются аномальные точки, включая запись с отрицательным числом покупок при наличии дохода (-1 покупка, 785 дохода) и случаи, когда покупки отсутствуют, но доход положителен.
